O objetivo deste notebook é mostrar como calcular o negativo de uma região dentro de uma imagem, desenhando tal região sobre a imagem original.

```python
# Importar bibliotecas necessárias
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Carregar imagem original
input_image = cv2.imread('arara2.jpg', cv2.IMREAD_GRAYSCALE)

plt.title('Input image')
plt.imshow(input_image,'gray')
plt.show()

print("Dimensões da imagem: ",input_image.shape)
```

## Negativo da imagem

Fórmula para o cálculo do negativo de uma imagem:

$$s = L-1-r$$

Onde:

- L é a quantidade de níveis de intensidade com a qual a imagem é representada;
- r é um valor de intensidade na imagem original
- s é um valor de intensidade na imagem pós processada

A implementação de tal função pode ser feita da seguinte forma:

```python
def inverter_intensidade(image, x_begin, x_end, y_begin, y_end, L=256):
    # Testar se os valores da região não ultrapassam os limites da imagem
    if x_end >= image.shape[0] or y_end >= image.shape[1]:
        print('Região escolhida ultrapassa limites da imagem')
        return 0
    else:
        # Percorrer a região da imagem escolhida e calcular o negativo dentro dela
        new_image = image.copy()
        for i in range(x_begin,x_end):
            for j in range(y_begin,y_end):
                new_image[i,j] = L-1-image[i,j]
        return new_image
```

Escolhe-se uma região retangular dentro da imagem, delimitada por pontos P1 e P2 do tipo (x,y):

```python
P1 = (90,200)
P2 = (500,620)
```

Passamos a imagem e os pontos para função como argumentos e armazenamos a nova imagem:

```python
negative_image = inverter_intensidade(input_image, P1[0], P2[0], P1[1], P2[1])
```

```python
fig,ax = plt.subplots(1,2)
plt.subplot(121),plt.imshow(input_image,'gray')
plt.subplot(122),plt.imshow(negative_image, 'gray')
plt.show()
```

É possível também mostrar a imagem fora do notebook com o comando:

```python
# Cria as janelas para exibir as imagens
cv2.namedWindow('Imagem original',cv2.WINDOW_AUTOSIZE)
cv2.namedWindow('Imagem pós processada',cv2.WINDOW_AUTOSIZE)

# Exibe as imagens até que alguma tecla seja pressionada
cv2.imshow('Imagem original',input_image)
cv2.imshow('Imagem pós processada',negative_image)
cv2.waitKey(0)
```
